```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('CO2_emission.csv')
data.head()
```

```python
data.dropna()
```

```python
data1=data.groupby(by=data['Region']).mean()
data1
```

```python
# a=data1.mean(axis=1)
# a
```

```python
data1['max CO2 emission']=data1.max(axis=1)
data1
```

```python
plt.figure(figsize=(20,7))       # this shows the maximum emission of the mean of the emission of CO2 regionwise
plt.bar(data1.index,data1['max CO2 emission'])
plt.grid(alpha=0.2)
plt.title('Max CO2 emission regionwise',size=26)
plt.xlabel('Regions',size=16)
plt.ylabel('CO2 emission',size=16)
plt.show()
```

Graph 2

```python
data_region=data['Region'].value_counts()
data_region
```

```python
s=[0,0,0,0,0,0,0.2]                       # Percentage of the country names falling in a particular region
plt.figure(figsize=(14,4))
plt.pie(data_region,autopct='%0.2f%%',labels=data_region.index,explode=s)
plt.title('Region vs Country Name count in that region',size=18)
plt.show()
```

```python
plt.figure(figsize=(18,4))
plt.hist(data['Region'])
plt.show()
```

```python
plt.bar(data['Country Name'][0:6],data['country_code'][0:6])
```

Graph 4- 

```python
plt.figure(figsize=(20,4))                  # Graph showing the CO2 emission in each year corresponding to the country name
plt.scatter(data['Country Name'][1:10],data['1990'][1:10],label='1990')
plt.scatter(data['Country Name'][1:10],data['1991'][1:10],label='1991')
plt.scatter(data['Country Name'][1:10],data['1992'][1:10],label='1992')
plt.plot(data['Country Name'][1:10],data['2019.1'][1:10],marker='*',markersize=10,label='2019.1')
plt.title('CO2 emission in each country',size=24)
plt.xlabel('COO2 emission',size=16)
plt.ylabel('Countries',size=16)
plt.grid(alpha=0.3)
plt.legend(shadow=1)
plt.show()
```
